## 1. Compare Sales by region for 2016 with 2015 using bar chart.

```python
sales_metric_by_region = sales.loc[:, ['Region', 'Sales2015', 'Sales2016']]
sales_metric_by_region = sales_metric_by_region.groupby(['Region'])[['Sales2015', 'Sales2016']]\
                                                                                               .agg('sum')\
                                                                                               .map(int)\
                                                                                               .reset_index()
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
#fig, ax = plt.subplots(figsize = (10.5, 4.5), linewidth=5, edgecolor="#04253a")
fig, ax = plt.subplots(figsize = (10.5, 4.5))
plt.setp(ax.spines.values(), lw=2, color='black')

b1 = ax.barh(sales_metric_by_region['Region'], 
             sales_metric_by_region['Sales2015'], 
             xerr=10**5.5,
             label = True, 
             edgecolor = 'black',
             color = 'lightcoral')

b2 = ax.barh(sales_metric_by_region['Region'], 
             sales_metric_by_region['Sales2016'], 
             left = sales_metric_by_region['Sales2015'], 
             xerr=10**5.5,
             label = True, 
             edgecolor = 'black',
             color = 'powderblue')

ax.legend(['2015', '2016'], shadow = True, edgecolor='black', title = 'Year')
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
ax.bar_label(b1, label_type = 'edge', fmt = '{:,.2f}', color = 'black')
ax.bar_label(b2, label_type = 'edge', fmt = '{:,.2f}', color = 'black')
plt.xlabel('\nSales', color='dimgrey', font='roboto', size=15)
plt.ylabel('Regions\n', color='dimgrey', font='roboto', size=15)
plt.title("Sales by region | Comparison by year (2015-16)\n", color='dimgrey', font='roboto', size=15)
plt.show()
```

## 2. What are the contributing factors to the sales for each region in 2016. Visualize it using a Pie Chart.

```python
sales_metric_by_region = sales.loc[:, ['Region', 'Sales2015', 'Sales2016']]
sales_metric_by_region = sales_metric_by_region.groupby(['Region'])[['Sales2015', 'Sales2016']]\
                                                                                               .agg('sum')\
                                                                                               .map(int)\
                                                                                               .reset_index()

Region = sales_metric_by_region['Region']
sales2016 = sales_metric_by_region['Sales2016']

fig, ax = plt.subplots(figsize = (8.5, 6.5))
ax.pie(sales2016, labels = Region, 
       shadow = True, 
       explode = (0.0, 0.1, 0.0), 
       colors = ['crimson', 'silver', 'beige'], 
       autopct = "%1.2f%%", textprops = dict(color='black'))

plt.title("Sales by region - 2016", color='dimgrey', font='roboto', size=15)
plt.show()
```

## 3. Compare the total sales of 2015 and 2016 with respect to Region and Tiers.

### Table view

```python
sales_metric_by_region2 = sales.loc[:, ['Region', 'Tier', 'Sales2015', 'Sales2016']]
sales_metric_by_region2 = sales_metric_by_region2.groupby(['Region', 'Tier'])[['Sales2015', 'Sales2016']]\
                                                                                               .agg('sum')\
                                                                                               .map(int)\
                                                                                               
sales_metric_by_region2.style\
                       .background_gradient(cmap = 'YlGnBu_r', 
                                            subset = ['Sales2015', 'Sales2016'])\
                       .highlight_between(left = -10000000, 
                                          right = 0, 
                                          subset = ['Sales2015', 'Sales2016'], 
                                          color = 'crimson')\
                       .format({"Sales2015" : "{:,.2f}", "Sales2016" : "{:,.2f}"})
```

### Graph view

```python
sales_metric_by_region2 = sales_metric_by_region2.reset_index()
regList = sales_metric_by_region2['Region'].unique()

for i in regList:
    plotVal = sales_metric_by_region2[sales_metric_by_region2['Region'] == i]
    sns.set_style("darkgrid", {"axes.facecolor": ".9"})
    fig, ax = plt.subplots(figsize=(8.5, 7.5))
    plt.setp(ax.spines.values(), lw=2, color='black')
    
    b1 = plt.bar(plotVal['Tier'], 
                 plotVal['Sales2015'], 
                 yerr=10**5.5, 
                 align = 'center', 
                 label = True, 
                 edgecolor = 'black', 
                 color = 'lightcoral')
    
    b2 = plt.bar( plotVal['Tier'], 
                 plotVal['Sales2016'], 
                 yerr=10**5.5, 
                 align = 'center', 
                 label = True, 
                 edgecolor = 'black', 
                 bottom = plotVal['Sales2015'], 
                 color = 'powderblue')
    
    plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
    plt.title("total sales in {0} region\n".format(i), color='dimgrey', font='roboto', size=15)
    ax.bar_label(b1, label_type = 'edge', fmt = '{:,.0f}')
    ax.bar_label(b2, label_type = 'edge', fmt = '{:,.0f}')
    plt.legend(['2015', '2016'], title='Year', shadow=True, edgecolor='black')
    plt.ylabel('\nSales', color='dimgrey', font='roboto', size=15)
    plt.xlabel('\nTiers', color='dimgrey', font='roboto', size=15)
    
    plt.show()
    print('\n\n')
```

## 6. Create a new column Qtr using numpy.where() or any suitable utility in the imported dataset. The Quarters are based on months and defined as -


**•  Jan - Mar : Q1**

**•  Apr - Jun : Q2**

**•  Jul - Sep : Q3**

**•  Oct - Dec : Q4**

### Step-1: Get the column to classify months into Quarter

```python
sales['Month Number']  =  pd.to_datetime(sales['Month'], format='%b')

sales['Month Number']  =  sales['Month Number'].dt.strftime('%m')

sales['Month Number'] = sales['Month Number'].apply(lambda x: int(x))

sales.loc[sales['Month Number'].between(1, 3, 'both'), 'Qtr'] = 'Q1'
sales.loc[sales['Month Number'].between(4, 6, 'both'), 'Qtr'] = 'Q2'
sales.loc[sales['Month Number'].between(7, 9, 'both'), 'Qtr'] = 'Q3'
sales.loc[sales['Month Number'].between(10, 12, 'both'), 'Qtr'] = 'Q4'

sales.drop('Month Number', axis = 'columns', inplace = True)
```

### Step-2: Check if number of records in "Qtr" columns are consistent

```python
Q_months = [{'Q1' : ['Jan', 'Feb', 'Mar']}, 
            {'Q2' : ['Apr', 'May', 'Jun']}, 
            {'Q3' : ['Jul', 'Aug', 'Sep']}, 
            {'Q4' : ['Oct', 'Nov', 'Dec']}]


for i in Q_months:
    for k, v in i.items():
        qtr_rec_cnt = len(sales.loc[sales['Qtr']==k, :])
        quarter_record_counter = qtr_rec_cnt
        months_record_counter = 0
        print("Count of records in - ", k , ": ", qtr_rec_cnt)
        for months_name in v:
            month_cnt = len(sales.loc[sales['Month']==months_name, :])
            months_record_counter += month_cnt
            print("Count of records for Month - ", months_name , ": ", months_record_counter)
           
        print("Total records from each given month:", months_record_counter)
        print('')
        print("Consistent with individual month record count?", months_record_counter==qtr_rec_cnt)
        print("\n")
```
